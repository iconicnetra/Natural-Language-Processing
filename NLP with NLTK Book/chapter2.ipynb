```python
from__future__import division
def lexical_diversity(text):
    return len(text)/len(text)
```

```python
def lexical_diversity(my_text_data):
    word_count = len(my_text_data)
    vocab_size = len(set(my_text_data))
    diversity_score = word_count/vocab_size
    return diversity_score
```

```python
def plural(word):
    if word.endswith('y'):
        return word[:-1] + 'ies'
    elif word[-1] in 'sx' or word[-2:] in ['sh', 'ch']:
        return word + 'es'
    elif word.endswith('an'):
        return word[:-2] + 'en'
    else:
        return word + 's'
```

# Lexical Resources

A lexicon, or lexical resource, is a collection of words and/or phrases along with asso
ciated information, such as part-of-speech and sense definitions. Lexical resources are
 secondary to texts, and are usually created and enriched with the help of texts.

## Wordlist Corpora

```python
nltk.download('words')
```

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab - english_vocab
    return sorted(unusual)
```

```python
unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))
```

## Corpus of StopWords

```python
from nltk.corpus import stopwords 
stopwords.words('english')
```

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content)/len(text)
```

```python
syllable = ['N', 'IH0', 'K', 'S']
[word for word, pron in entries if pron[-4:] == syllable]
```

```python
sorted(set(w[:2] for w, pron in entries if pron[0] == 'N' and w[0] != 'n'))
```

```python
def stress(pron):
    return [char for phone in pron for char in phone if char.isdigit()]
[w for w, pron in entries if stress(pron)==['0', '1', '0', '2', '0']] #Finding the words that have the stress pattern 0-1-0-2-0
```

```python
p3 = [(pron[0]+'-'+pron[2], word) for word, pron in entries if pron[0] == 'P' and len(pron) == 3]
cfd = nltk.ConditionalFreqDist(p3)
for template in cfd.conditions():
    if len(cfd[template]) > 10:
        words = cfd[template].keys()
        wordlist = ' '.join(words)
        print(template, wordlist[:70] + "...")
```
