```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")  

doc = nlp("The quick brown fox jumps over the lazy dog. It was a bright, sunny day and the park was full of cheerful people. Children were playing near the fountain, their laughter echoing through the air. Nearby, a group of friends were having a picnic under a large oak tree. A couple was walking their dog along the path, enjoying the beautiful weather. In the distance, the city skyline could be seen against the clear blue sky.")

for token in doc:
    print(token.text, token.pos_)
```

```python
for token in doc:
    print(token.text, token.dep_)
```

```python
for token in doc:
    print(token, " | ", token.pos_, " | ", token.dep_)
```

```python
for token in doc:
    print(token, " | ", token.pos_)
```

```python
for token in doc:
    print(token, " | ", token.pos_, " | ",spacy.explain(token.pos_))
```

```python
doc = nlp("Wow! Dr. Strange made 400 million dollars in the box office on the first day.")

for token in doc:
    print(token, " | ", token.pos_, " | ",spacy.explain(token.pos_))
```

```python
para = """Once upon a time, in a land far away, there was a brave knight named Sir Lancelot. 
He rode his trusty steed through the dense forest, his silver armor glistening in the moonlight. 
His mission was to rescue the princess who was held captive in the dragon's lair. Despite the perilous journey ahead, 
Sir Lancelot was determined and fearless. 
He knew that he had to face the fire-breathing dragon and overcome numerous challenges to save the princess and bring peace to the kingdom."""

paragraph = nlp(para)

for token in paragraph:
    print(token, " | ", token.pos_, " | ",spacy.explain(token.pos_))
```

```python
paragraph = nlp(para)

for token in paragraph:
    print(token, " | ", token.pos_, " | ",spacy.explain(token.pos_), " | ", token.dep_, " | ", spacy.explain(token.dep_))
```

```python
paragraph = nlp(para)

for token in paragraph:
    print(token, " | ", token.pos_, " | ",spacy.explain(token.pos_), " | ", token.tag_, " | ", spacy.explain(token.tag_))
```

## Removing all Spaces, Punct, and X tokens from the Text 

```python
earnings_text="""Microsoft Corp. today announced the following results for the quarter ended December 31, 2021, as compared to the corresponding period of last fiscal year:

·         Revenue was $51.7 billion and increased 20%
·         Operating income was $22.2 billion and increased 24%
·         Net income was $18.8 billion and increased 21%
·         Diluted earnings per share was $2.48 and increased 22%
“Digital technology is the most malleable resource at the world’s disposal to overcome constraints and reimagine everyday work and life,” said Satya Nadella, chairman and chief executive officer of Microsoft. “As tech as a percentage of global GDP continues to increase, we are innovating and investing across diverse and growing markets, with a common underlying technology stack and an operating model that reinforces a common strategy, culture, and sense of purpose.”
“Solid commercial execution, represented by strong bookings growth driven by long-term Azure commitments, increased Microsoft Cloud revenue to $22.1 billion, up 32% year over year” said Amy Hood, executive vice president and chief financial officer of Microsoft."""

doc = nlp(earnings_text)

filtered_tokens = []

for token in doc:
    if token.pos_ not in ["PUNCT", "SPACE", "X"]:
        filtered_tokens.append(token)
```
